Matplotlib

```
import numpy as np
import matplotlib.pyplot as plt
```

```
plt.plot([1,7,9,14])
```

```
plt.plot([1,2,7,8], [1,9,7,14])
```

```
plt.plot([1,2,7,8], [1,9,7,14],'ro')
```

```
a = np.array([[1,2,3,4],
              [5,6,7,8]])
plt.plot(a[0],a[1],'go')
```

```
plt.xlabel('x-axis')
plt.ylabel('y-axis')

plt.plot(a[0],a[1],'b--')
```

```
t = np.linspace(0, 5, 20)

plt.plot(t, t**2, color='coral', linestyle='--')
plt.plot(t, 4*t, color= 'indigo',linewidth = 2)
```

```
plt.subplot(1,3,1)
plt.plot(t,t,'c--')
plt.subplot(1,3,2)
plt.plot(t,t**2,'b-')
plt.subplot(1,3,3)
plt.plot(t,t**3,'mo')
```

```
grades = {'Pulkit': 40,
         'Sourav': 41,
         'Jigar': 42}

plt.bar(range(len(grades)), list(grades.values()))
plt.xticks(range(len(grades)), list(grades.keys()))
plt.xlabel('Students')
plt.ylabel('Grades')
```
